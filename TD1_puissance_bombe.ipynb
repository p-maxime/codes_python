```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

t = np.array([0.1,0.2,0.4,0.5,0.7,0.8,0.9,1.1,1.2,1.4,1.5,1.7,1.8,1.9,3.3,3.5,3.8,4.1,4.3,4.6,15,25,34,53,62])
r = np.array([11.1,19.9,25.4,28.8,31.9,34.2,36.3,38.9,41.0,42.8,44.4,46.0,46.9,48.7,59,61.1,62.9,64.3,65.6,67.3,106.5,130,145,175,185])
s, e = 0, np.size(t)

#fit_values, *fit_residuals = np.polyfit(t[s:e]**2,r[s:e]**4,1,full = True)
#print(fit_values, fit_residuals)
#fit_model = np.poly1d(fit_values)
#plt.plot(t**2,fit_model(t**2))

#fonction de mod√©lisation
def lin(x,a):
    return a*x
#Mod√©lisation
popt, copt = curve_fit(lin,t[s:e]**2,r[s:e]**4)
fit_model = popt[0]*t**2
print()
plt.figure(figsize=(10,6))
plt.plot(t**2,r**4,'+')
plt.plot(t**2,fit_model,label='E = {:.3e} J; u(E) = {:.1e} J'.format(popt[0]*1.2*1e6,np.sqrt(copt[0][0])*1.2*1e6))#popt[0] est une estimation de c*E/rho avec c constante suppos√© = 1 et rho = 1,2 kg/m^3 et t en ms => 1e6 
plt.xlabel('$t^2$ avec $t$ en ms')
plt.ylabel('${r}^4$ avec $r$ en m')
plt.legend()
plt.show()

```
