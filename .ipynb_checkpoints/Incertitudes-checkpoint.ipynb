### Mesure des incertitudes des traits d'une règle

première partie importer l'image, la découper et placer les lignes 

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

def crop(img,lmin,lmax,cmin,cmax):
    return img[lmin:lmax,cmin:cmax,0]

%matplotlib notebook


img = mpimg.imread('im4.jpg')
plt.figure(figsize=(10,10))
imgcrop = crop(img,1400,5000,1000,2000)
plt.imshow(img[:,:,0])
plt.show()

plt.figure(figsize=(10,10))
imgcrop = crop(img,500,2500,0,4000)
imgcrop = imgcrop.transpose()
plt.imshow(imgcrop)
c1 = 700
c2 = 1150
plt.axvline(c1,color = 'red')
plt.axvline(c2,color = 'blue')
plt.show()
```

```python

r1 = img[:,c1,0] #rouge
r2 = img[:,c2,0] #bleu
r3 = imgcrop[:,c1] #rouge
r4 = imgcrop[:,c2] #bleu

def trigger(r, s, f = 1, e = 1, p = 8):
    """
    r : vecteur ou liste
    s : valeur du seuil de déclenchement
    f : front montant 1 (valeur par défaut) front descendand -1
    e : écart en pixel pour passer la valeur seuil 
    p : pause en pixel avant le nouveau déclenchement (pour un comportement normal e doit être < à p)
    """
    L=[]
    pause = 0
    r = np.array(r,dtype='int16')#attention à bien convertir uint8 en int16 pour les valeurs négatives 
    for i in range(len(r)-p):
        if pause == 0:
            if r[i]*f <= s*f and r[i+e]*f > s*f:
                ind, = np.where(np.abs(r[i:i+p]-s) == np.min(np.abs(r[i:i+p]-s))) 
                L.append(i+ind[0])
                pause = p 
        else:
            pause-=1        
    return L

seuil = 100
r = r4
L = trigger(r,seuil)  

plt.figure(figsize=(10,10))
plt.plot(r)
plt.axhline(seuil,color="red")
plt.plot(L,r[L],'+',color="purple")
plt.show()


plt.figure(figsize=(10,6))
plt.subplot(1,2,1)
ecarts = np.diff(L)
plt.hist(ecarts, bins=range(ecarts.min(),ecarts.max()+2), label="Moyenne = {:.1f} px \nÉcart type = {:.1f} px".format(np.mean(np.diff(L)),np.std(np.diff(L))))
plt.legend()
plt.subplot(1,2,2)
plt.plot(np.diff(L),'+')
plt.show()

#L2 = L
#L3 = L 
#L4 = L

```

```python
plt.figure(figsize=(10,10))
plt.plot([L2[x]-L4[x] for x in range(len(L2))])
plt.show()
```
