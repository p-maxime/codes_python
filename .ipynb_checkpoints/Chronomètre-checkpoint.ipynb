Importation des bibliothèques :

```python
import numpy as np
import numpy.random as rd
import pandas as pd
import datetime 
import matplotlib.pyplot as plt
```

Importation du fichier 

Renommez le fichier de mesure en "chrono_prenom.txt" puis modifiez le texte "chrono_1.txt" en "chrono_prenom.txt"

```python
File = pd.read_csv(r'chrono_1.txt', delimiter="\t", header = None) # récupération du fichier txt (3ème colonne temps)
#File = pd.read_csv(r'.\chronometre\chrono_2s_b.txt', delimiter="\t", header = None) # récupération du fichier txt (3ème colonne temps)
temps = File.to_numpy() # fichier converti en tableau de chaîne de caractères (format "hh:mm:ss.xxx")

Dt = [] #Liste des intervalles de temps mesurées toutes les 2 secondes    
for i in range(np.size(temps,0)-1):
    t = temps[i,1] #récupération de la colonne des temps en format "hh:mm:ss.xxx"
    # conversion en millisecondes : 
    # des minutes : mm*60*1000
    # des secondes : + ss*1000
    # des millisecondes : + xxx
    Dt.append(float(t[3:5])*60*1000+float(t[6:8])*1000+float(t[9:13]))     
```

Mesure de l'intervalle de temps moyen et écart type de ce processus de mesure 

```python
Dtmoy = np.mean(Dt)
ecart_type = np.std(Dt,ddof=1)
#print("""La mesure manuelle au chronomètre d'une période de 2s donne une valeur moyenne de l'intervalle de \n Dtmoy = {:.0f} ms ; $u(Dtmoy) = {:.0f} ms""".format(Dtmoy,ecart_type/np.sqrt(len(Dt))))
```

Tracé des différentes mesures 

```python
plt.figure(figsize = (10,6)) 
plt.plot((Dt,Dt),(0,1), 'blue')
plt.plot((Dtmoy,Dtmoy),(0,1), 'red', lw=4)
plt.annotate('',xy=(Dtmoy-ecart_type,0.5),xytext=(Dtmoy,0.5), arrowprops=dict(arrowstyle='<->',lw=2))
plt.annotate('',xy=(Dtmoy,0.5),xytext=(Dtmoy+ecart_type,0.5), arrowprops=dict(arrowstyle='<->',lw=2))
plt.show()
```

Le trai rouge représente la valeur moyenne ``Dt_moy`` de votre processus de mesure et chaque trai bleu une mesure unique.
Les flèches représentent l'écart type ``ecart_type`` (une estimation de sa valeur) qui correspond à la formule :
$$\sigma = \sqrt{\dfrac{\sum_{n=1}^N(x_i-\overline{x})^2}{N-1}}$$
où $\sigma$ est ``ecart_type``, $\overline{x}$ est ``Dt_moy``, $t_i$ correspond à la i+1-ème valeur de la liste ``Dt`` et N est la taille de la liste ``Dt``.
Une autre représentation est l'histogramme, ce qui permet de mieux visualiser l'ensemble de vos mesures :

```python
#tracé d'un histogramme de la différence des temps (doit être normalement centré sur 0) 
plt.figure(figsize = (10,6))    
num, bins, patches = plt.hist(Dt,range=(1500,2500),bins=30,label='Moyenne = {:.1f} ms, Écart type = {:.0f} ms'.format(np.mean(Dt),np.std(Dt,ddof=1)))
plt.plot((Dtmoy,Dtmoy),(0,max(num)),'red',lw=4)
plt.annotate('',xy=(Dtmoy-ecart_type,max(num)*0.61),xytext=(Dtmoy,max(num)*0.61), arrowprops=dict(arrowstyle='<->', lw=2))
plt.annotate('',xy=(Dtmoy,max(num)*0.61),xytext=(Dtmoy+ecart_type,max(num)*0.61), arrowprops=dict(arrowstyle='<->', lw=2))
plt.xlabel('Ecart de temps (millisecondes)')
plt.ylabel('Nombre des écarts de temps situé dans un intervalle donné')
plt.legend()
plt.show()
```

### Génerer l'équivalent de 100000 mesures de temps
Vous réalisez maintenant 100000 mesures. Un peu long ? La fonction ci-dessous permet de générer un ensemble de mesure tel que la valeur moyenne soit ``Dtmoy`` et l'écart type ``ecart_type``. 

```python
Dt_MC = rd.normal(Dtmoy,ecart_type,100000) #ddof : degrés de liberté => calcul du std avec "N-1"
Dtmoy_MC = np.mean(Dt_MC)
ecart_type_MC = np.std(Dt_MC,ddof=1)
plt.figure(figsize=(12,6))   
num, bins, patches = plt.hist(Dt_MC,range=(1500,2500),bins=40,label='Moyenne = {:.1f}, Std = {:.0f}'.format(Dtmoy_MC,ecart_type_MC))
plt.plot((Dtmoy_MC,Dtmoy_MC),(0,max(num)),'red',lw=4)
plt.annotate('',xy=(Dtmoy_MC-ecart_type_MC,max(num)*0.61),xytext=(Dtmoy_MC,max(num)*0.61), arrowprops=dict(arrowstyle='<->', lw=2))
plt.annotate('',xy=(Dtmoy_MC,max(num)*0.61),xytext=(Dtmoy_MC+ecart_type_MC,max(num)*0.61), arrowprops=dict(arrowstyle='<->', lw=2))
plt.xlabel('Ecart de temps (millisecondes)')
plt.ylabel('Nombre des écarts de temps situé dans un intervalle donné')
plt.legend()
plt.show()
```
