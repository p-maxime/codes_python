***Exercice 5***  
Tapez le code suivant :
```
print(sin(90))
```
La fonction sinus n'est pas reconnue par Python. Il faut importer cette fonction à l'aide d'une bibliothèque. Il en existe plusieurs bibliothèques disponibles comportant cette fonction comme "math" ou "numpy". Vous pouvez soit importer la fonction désirée à l'aide de la commande :
```
from math import sin

print(sin(90))
```
soit importer l'ensemble des fonctions contenues dans la bibliothèque 

```
import numpy

print(numpy.sin(90))
a = numpy.radians(90)
print(numpy.sin(a))
```

l'appel de la fonction se fait par "bibliothèque.nom_fonction". Pour éviter les noms trops longs, il est possible d'utiliser une étiquette pour nommer la bibliothèque :

```
import numpy as np

print(np.sin(90))
```

Nous utiliserons toujours cette dernière notation.

***Exercice 8***  
Il est possible d'utiliser plusieurs bibliothèques en même temps. Par exemple, créer des histogrammes avec matplotlib.pyplot de nombres aléatoires distribués selon une "loi gaussienne" avec numpy.random.normal(centre,écart type,nombre de valeurs).
```
import numpy as np
import matplotlib.pyplot as plt

plt.figure()
plt.hist(np.random.normal(0,1,1000), bins = 50)
plt.draw()
```
