***Exercice 5***  
Tapez le code suivant :
```
print(sin(90))
```
La fonction sinus n'est pas reconnue par Python. Il faut importer cette fonction  l'aide d'une bibliothque. Il en existe plusieurs bibliothques disponibles comportant cette fonction comme "math" ou "numpy". Vous pouvez soit importer la fonction dsire  l'aide de la commande :
```
from math import sin

print(sin(90))
```
soit importer l'ensemble des fonctions contenues dans la bibliothque 

```
import numpy

print(numpy.sin(90))
a = numpy.radians(90)
print(numpy.sin(a))
```

l'appel de la fonction se fait par "bibliothque.nom_fonction". Pour viter les noms trops longs, il est possible d'utiliser une tiquette pour nommer la bibliothque :

```
import numpy as np

print(np.sin(90))
```

Nous utiliserons toujours cette dernire notation.
